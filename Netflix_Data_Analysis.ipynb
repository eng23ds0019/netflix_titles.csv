```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")

print("Path to dataset files:", path)
import pandas as pd

# Load the Netflix dataset from the downloaded path
file_path = "/kaggle/input/netflix-shows/netflix_titles.csv"

# Read the dataset
df = pd.read_csv(file_path)

# Show the first few rows
df.head()
# Check the number of rows and columns
df.shape
# List all column names
df.columns
# Check the data types of each column
df.dtypes
# Summary of dataset (non-null counts, types)
df.info()
# Check how many null values are in each column
df.isnull().sum()
# Summary statistics (only for numerical values)
df.describe()
# See unique values in a specific column like 'type'
df['type'].value_counts()
# Drop duplicate rows (if any)
df.drop_duplicates(inplace=True)

# Drop rows with missing values (only if they are few)
df.dropna(subset=['title', 'type', 'country'], inplace=True)
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='type', palette='viridis')
plt.title("Distribution of Content Type on Netflix")
plt.xlabel("Type")
plt.ylabel("Count")
plt.show()
# Count top countries
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='magma')
plt.title("Top 10 Countries with Most Netflix Titles")
plt.xlabel("Number of Titles")
plt.ylabel("Country")
plt.show()
# Convert 'date_added' to datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract the year
df['year_added'] = df['date_added'].dt.year

# Count per year
yearly_counts = df['year_added'].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o')
plt.title("Netflix Titles Added per Year")
plt.xlabel("Year")
plt.ylabel("Number of Titles Added")
plt.grid(True)
plt.show()
from collections import Counter

# Split all genres and flatten them
genre_series = df['listed_in'].dropna().apply(lambda x: x.split(', '))
all_genres = [genre for sublist in genre_series for genre in sublist]

# Count frequency of each genre
genre_counts = Counter(all_genres)

# Convert to DataFrame
genre_df = pd.DataFrame.from_dict(genre_counts, orient='index', columns=['count']).sort_values(by='count', ascending=False)

# Plot top 10 genres
plt.figure(figsize=(10,5))
sns.barplot(x=genre_df.head(10).index, y=genre_df.head(10)['count'], palette="coolwarm")
plt.xticks(rotation=45)
plt.title("Top 10 Genres on Netflix")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
# Drop nulls and count appearances of each country
country_series = df['country'].dropna().apply(lambda x: x.split(', '))
all_countries = [country.strip() for sublist in country_series for country in sublist]

from collections import Counter
country_counts = Counter(all_countries)

# Convert to DataFrame
country_df = pd.DataFrame.from_dict(country_counts, orient='index', columns=['count']).sort_values(by='count', ascending=False)

# Plot top 10 countries
plt.figure(figsize=(10,5))
sns.barplot(x=country_df.head(10).index, y=country_df.head(10)['count'], palette="viridis")
plt.xticks(rotation=45)
plt.title("Top 10 Countries with Most Netflix Content")
plt.xlabel("Country")
plt.ylabel("Number of Titles")
plt.tight_layout()
plt.show()
# Drop missing values in 'release_year' column
df['release_year'] = df['release_year'].dropna()

# Count releases by year
release_trend = df['release_year'].value_counts().sort_index()

# Plot the trend
plt.figure(figsize=(12, 6))
sns.lineplot(x=release_trend.index, y=release_trend.values, marker='o')
plt.title('Year-wise Trend of Content Release on Netflix')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
# Count the number of Movies vs TV Shows
content_type = df['type'].value_counts()

# Plotting
plt.figure(figsize=(6, 6))
plt.pie(content_type, labels=content_type.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=140)
plt.title('Distribution of Content Types on Netflix')
plt.axis('equal')
plt.show()
# Drop rows with missing director names
df_directors = df.dropna(subset=['director'])

# Count the number of titles per director
top_directors = df_directors['director'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
# Drop rows with missing 'country' values
df_country = df.dropna(subset=['country'])

# Extract the first country in case multiple are listed
df_country['main_country'] = df_country['country'].apply(lambda x: x.split(',')[0])

# Count titles per country
top_countries = df_country['main_country'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='plasma')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
# Remove nulls in country
df_country = df.dropna(subset=['country'])

# Use only the first country if multiple listed
df_country['main_country'] = df_country['country'].apply(lambda x: x.split(',')[0])

# Count top 10 countries
top_countries = df_country['main_country'].value_counts().head(10)

# Plot it
plt.figure(figsize=(10, 5))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='plasma')
plt.title('Top 10 Countries with Most Netflix Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.show()
# Extract year from date_added
df['year_added'] = pd.DatetimeIndex(df['date_added']).year

# Count content per year
content_per_year = df['year_added'].value_counts().sort_index()

# Plot
plt.figure(figsize=(10, 5))
sns.lineplot(x=content_per_year.index, y=content_per_year.values, marker='o')
plt.title('Netflix Content Added Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()
# Group by year and type
type_year = df.groupby(['year_added', 'type']).size().unstack()

# Plot
type_year.plot(kind='bar', stacked=False, figsize=(12, 6), colormap='Set1')
plt.title('Movies vs TV Shows Added Each Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend(title='Type')
plt.tight_layout()
plt.show()
# View common durations
plt.figure(figsize=(12, 5))
sns.countplot(data=df[df['type'] == 'Movie'], x='duration', order=df['duration'].value_counts().iloc[:10].index, palette='inferno')
plt.title('Top 10 Most Common Movie Durations')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
# Drop nulls from director
df_directors = df.dropna(subset=['director'])

# Count top 10 directors
top_directors = df_directors['director'].value_counts().head(10)

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='magma')
plt.title('Top 10 Directors on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.tight_layout()
plt.show()
# Drop missing ratings
df_ratings = df.dropna(subset=['rating'])

# Count and plot
plt.figure(figsize=(10, 5))
sns.countplot(data=df_ratings, x='rating', order=df_ratings['rating'].value_counts().index, palette='coolwarm')
plt.title('Distribution of Netflix Content Ratings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
from wordcloud import WordCloud

# Combine all titles
title_words = ' '.join(df['title'])

# Generate word cloud
wordcloud = WordCloud(width=1000, height=500, background_color='black', colormap='spring').generate(title_words)

# Show plot
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Most Frequent Words in Netflix Titles', fontsize=20)
plt.show()
# Add title length column
df['title_length'] = df['title'].apply(len)

# Get top 10 longest titles
long_titles = df[['title', 'title_length']].sort_values(by='title_length', ascending=False).head(10)
print("Top 10 Longest Titles on Netflix:")
print(long_titles)
# Drop nulls from cast
df_cast = df.dropna(subset=['cast'])

# Combine all actor names into one list
from collections import Counter
actors = df_cast['cast'].str.split(', ').sum()
top_actors = Counter(actors).most_common(10)

# Convert to DataFrame
actor_df = pd.DataFrame(top_actors, columns=['Actor', 'Appearances'])

# Plot
plt.figure(figsize=(10, 5))
sns.barplot(data=actor_df, x='Appearances', y='Actor', palette='cubehelix')
plt.title('Top 10 Most Frequent Actors on Netflix')
plt.tight_layout()
plt.show()
from collections import Counter

# Split and clean actor names
actor_list = df['cast'].dropna().str.split(', ')
all_actors = [actor for sublist in actor_list for actor in sublist]

# Count appearances
actor_count = Counter(all_actors)
top_actors = actor_count.most_common(10)

# Convert to DataFrame for plotting
actor_df = pd.DataFrame(top_actors, columns=['Actor', 'Count'])

# Plot
sns.barplot(x='Count', y='Actor', data=actor_df)
plt.title('Top 10 Actors on Netflix')
plt.xlabel('Number of Shows')
plt.ylabel('Actor')
plt.show()
# Prepare data
genre_type = df[['type', 'listed_in']].copy()
genre_type['listed_in'] = genre_type['listed_in'].str.split(', ')
genre_type = genre_type.explode('listed_in')
genre_count = genre_type.groupby(['listed_in', 'type']).size().unstack().fillna(0)

# Plot
genre_count.head(10).plot(kind='bar', stacked=True, figsize=(12,6))
plt.title('Top Genres by Content Type')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
